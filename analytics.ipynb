```python
import pandas as pd

df = pd.read_csv('Mock Data/Mock Data Lec 4.csv')
df.drop(columns = ["Unnamed: 13"], inplace=True)
df.head()

```

## Data Analytics for Preliminary Grades

```python
# Create lists for each range
prelim_range1 = []  # Grades 0 to 59
prelim_range2 = []  # Grades 60 to 69
prelim_range3 = []  # Grades 70 to 79
prelim_range4 = []  # Grades 80 to 89
prelim_range5 = []  # Grades 90 to 100

# Categorization loop
for grade in df['PRELIM GRADE']:
    if 0 < grade <= 59:
        prelim_range1.append(grade)
    elif 60 < grade <= 69:
        prelim_range2.append(grade)
    elif 70 < grade <= 79:
        prelim_range3.append(grade)
    elif 80 < grade <= 89:
        prelim_range4.append(grade)
    elif 90 < grade <= 100:
        prelim_range5.append(grade)

# Count students in each range
count1_range1 = len(prelim_range1) # Counting the number of students for this score range 0 to 59
count1_range2 = len(prelim_range2) # Counting the number of students for this score range 60 to 69
count1_range3 = len(prelim_range3) # Counting the number of students for this score range 70 to 79
count1_range4 = len(prelim_range4) # Counting the number of students for this score range 80 to 89
count1_range5 = len(prelim_range5) # Counting the number of students for this score range 90 to 100

result1 = f"""With the total of {count1_range1}, the range 0–59 has the highest number of students. It indicates a significant portion of the class is struggling to meet even the minimum expected learning outcomes. 
This suggests an urgent need to address student learning habits, such as poor time management, lack of study strategies, and low motivation. However, these challenges are often compounded by instructional methods that may not effectively engage or accommodate diverse learning styles. 
Therefore, it is recommended that instructors employ more formative assessments, incorporate diagnostic feedback, and use active learning strategies to foster student engagement and conceptual understanding."""

result2 = f"""Having a total of {count1_range2} In the 60–69 range, students are passing, but only marginally. This suggests that while basic comprehension may be present, there are still substantial gaps in critical thinking and application of concepts. 
The issue may lie in surface-level learning, where students rely on memorization rather than deep understanding. Here, both student habits and curricular clarity need to be improved. 
Educators should emphasize reflective learning strategies, provide explicit rubrics, and guide students toward self-regulation and metacognition. 
Curriculum designers should also revisit the alignment between intended learning outcomes, instructional strategies, and assessments to ensure coherence.
"""

result3 = f"""A total of {count1_range3} students obtained scores within the 70–79 range represent a “good” level of understanding. Most students here demonstrate competence but might not be challenged enough to progress toward mastery. 
Therefore, the curriculum may require enhancement to better stimulate critical thinking and real-world application. The inclusion of problem-based learning, collaborative tasks, and case studies can significantly improve engagement and depth of learning.
At this level, teaching strategies should shift from simply delivering content to cultivating a more inquiry-based, analytical approach.
"""

result4 = f"""A total of {count1_range4} students obtained a score within the 80–89 range, which indicates strong academic performance and effective learning. The current teaching strategies and curriculum structure appear effective, though minor enhancements could further stimulate these learners. 
In this case, advanced tasks, such as mini-research projects, debates, or leadership roles in group activities, are recommended to help these students reach their full potential. These learners can also be involved in peer tutoring, which benefits both their own learning and the overall class performance."""

result5 = f"""The {count1_range5} students scoring in the 90–100 range are excelling and potentially under-challenged by the current curriculum. These students would benefit from curriculum enrichment through experiential learning opportunities, independent research, and complex real-world scenarios. 
Instructors should consider their feedback to refine course content and ensure it remains stimulating. Encouraging these students to participate in academic competitions or lead class discussions can further refine their skills while contributing to a high-achieving academic environment."""


# Finding which range has the highest number of students
highest_range = max(count1_range1, count1_range2, count1_range3, count1_range4, count1_range5)

# Determine the corresponding range
if highest_range == count1_range1:
    print(result1)

elif highest_range == count1_range2:
    print(result2)

elif highest_range == count1_range3:
    print(result3)

elif highest_range == count1_range4:
    print(result4)

elif highest_range == count1_range5:
    print(result5)
    
```
