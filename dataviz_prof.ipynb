## A. Functional Suitability
**Directions:** Please evaluate the software quality of the NextGenCQI based on the characteristics defined by the ISO 25010 Software Product Quality Standards for end users. Use the five-point scale provided below to rate each characteristic, with 5 being the highest and 1 being the lowest. 

Scale: 5- Strongly Agree, 4- Agree, 3- Neutral, 2- Disagree, 1- Strongly Disagree

```python
# Your data
func_cmplt = {'1 - Strongly Disagree': 0, '2 - Disagree': 0, '3 - Neutral': 1, '4 - Agree': 1, '5 - Strongly Agree': 1}
func_corr  = {'1 - Strongly Disagree': 0, '2 - Disagree': 0, '3 - Neutral': 0, '4 - Agree': 2, '5 - Strongly Agree': 1}
func_appr  = {'1 - Strongly Disagree': 0, '2 - Disagree': 0, '3 - Neutral': 0, '4 - Agree': 1, '5 - Strongly Agree': 2}

rating = list(func_cmplt.keys())
x = np.arange(len(rating))  # X locations for the groups

# Values
cmplt_values = list(func_cmplt.values())
corr_values = list(func_corr.values())
appr_values = list(func_appr.values())

# Charting
style.use('ggplot')
plt.figure(figsize=(12, 5))

width = 0.25  # width of each bar

# Plotting
plt.bar(x - width, cmplt_values, width=width, label='Functional Completeness', color='#EA7300')
plt.bar(x, corr_values, width=width, label='Functional Correctness', color='#E83F25')
plt.bar(x + width, appr_values, width=width, label='Functional Appropriateness', color='#A62C2C')

# Details
plt.xlabel('System Rating')
plt.ylabel('Number of Responses')
plt.title('Functional Properties Comparison')
plt.xticks(x, rating)  # set x-axis labels
plt.legend()
plt.show()
```
